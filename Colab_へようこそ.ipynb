```
# ==============================
# Dolphin3.0 (Llama3.1-8B) 簡易推論ノートブック
# ==============================

# 1. 必要ライブラリのインストール
!pip install -q transformers accelerate bitsandbytes
```

```
# 2. モデルとトークナイザのロード
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

model_id = "cognitivecomputations/Dolphin3.0-Llama3.1-8B"
```

```
# トークナイザ
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```
# モデル (8bit 量子化で省メモリ動作)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    torch_dtype="auto",
    load_in_8bit=True
)
```

```

# 推論用パイプライン
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    device=0
)
```

# 3. 推論実行

```

prompt = "江戸時代の城下町の暮らしを子供向けに説明してください。"

output = pipe(
    prompt,
    max_new_tokens=200,
    temperature=0.7,
    top_p=0.9,
    do_sample=True
)

print("=== 応答 ===")
print(output[0]["generated_text"])
```
